## TZ Survey Analysis

```python
# import necessary libraries
import os
import pandas as pd
```

```python
# The first step is to fetch all the data

# determine the relative path of the folder holding the csv files
folder_path = "./TZA_2020_NPS-R5_v02_M_CSV"
```

```python
# Initialize an empty dictionary to store the data
data_dict = {}

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        # Create the variable name (removing the .csv extension)
        var_name = os.path.splitext(filename)[0]
        
        # Construct the full file path
        file_path = os.path.join(folder_path, filename)
        
        # Read the CSV file into a DataFrame
        data_dict[var_name] = pd.read_csv(file_path)
```

```python
# All the 67 datasets will need to be merged before we can explore which data is useful or not
# From tznps_2020_2021_basic_information_document_final.pdf (pg.15 - 24), we can get the unique identifier of each dataset
# we can use that to loop through the entire dataset and merge all results as appropriate
# to reduce chances of error, we will keep all merges as left merges, and we will first run the fn that does 
# single column merges, then double column merges


# Alternatively, we could determine the important datasets and merge only those.

# Whatever the case, we will need to have a function that takes an initial df and a list of whatever other number 
# of dfs and does the joins/merges to the end

#data_dict['hh_sec_a'].merge(hh_sec_b, how = 'left',  )
```

```python
def single_column_merger(left_df, right_dfs: list, merging_field: str):
    """Function receives a df, and a list of dfs and merges them
    These dfs must all have a common field on which to merge, which must be a string 
    (2 asserts here: common field, named as a string ) """
    # Create a list of suffixes to avoid a MergeError
    suffix_variables = ['df_'+str(i) for i in range(len(right_dfs))]
    print('suffix_variables is ', suffix_variables)
    for i in right_dfs:
        print(len(i))
        left_df = left_df.merge(i, how = 'left', on = merging_field, suffixes = suffix_variables)
    
    return left_df
    
```
